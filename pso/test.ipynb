```python
goal = lambda x, y: x**2+y**2 - x - y
```

```python
swarm = Swarm(15, 2, {"low" : [-1, -1], "high" : [1, 1]}, {
    "inertia_weight" : 0.3,
    "cognitive_coeff" : 0.8,
    "social_coeff" : 1.5,
}, plot=True)
```

```python
swarm.optimize(goal, 15)
```
